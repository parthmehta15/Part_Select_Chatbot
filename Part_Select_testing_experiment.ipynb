```python
# Send a request to the website
response = requests.get(dishwasher_url)
response
```

```python
# Parse the HTML content
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
fridge_brands = soup.find('ul', class_='nf__links').find_all('a')
fridge_brands_dict = {}
for link in fridge_brands:
    fridge_brands_dict[link.text] = main_url + link['href']
fridge_brands_dict
```

```python
fridge_parts_ul = soup.find('h2', id='ShopByPartType').find_next('ul')
fridge_parts_dict = {}
# Extract href and text from the 'ul' under "Related Refrigerator Parts"
for link in fridge_parts_ul.find_all('a'):
    fridge_parts_dict[link.text] = main_url + link['href']

fridge_parts_dict
```

Now to Loop through all the brand links and get list of parts by brand

```python
#Finding Secondary part links
fridge_secondary_parts_dict = {}
for name,url in fridge_brands_dict.items():
  print(name)
  response = requests.get(url)
  soup = BeautifulSoup(response.text, 'html.parser')
  try:
    fridge_secondary_parts_ul = soup.find('h2', id='ShopByPartType').find_next('ul')
  except:
    continue
  for link in fridge_secondary_parts_ul.find_all('a'):
    fridge_secondary_parts_dict[link.text] = main_url + link['href']

fridge_secondary_parts_dict
```

Now we have 3 dicts with all the parts:

fridge_brands_dict-> By going through each brand

fridge_parts_dict -> By going through each part

fridge_secondary_parts_dict -> Parts by brands by going through each brand

```python
refrigerator_issue_links = []
```

```python
for section in issue_sections:
        link = section['href']

        refrigerator_issue_links.append(f"https://www.partselect.com{link}")

refrigerator_issue_links
```

```python
refrigerator_issues_data = ''

for i in refrigerator_issue_links:
    print(i)
    response = requests.get(i)
    soup = BeautifulSoup(response.text, 'html.parser')

    #Find Relevent Content
    title = soup.find('h1', class_="title-main mt-3 mb-3 mb-lg-4>").text
    issues = soup.find_all('h2', class_="section-title bold col mt-3 mb-3")
    issues_desc = soup.find_all('div', class_="symptom-list__desc row mb-4")

    #Filter Content
    temp_data = ''''

    Issue: {title}

    Reasons:

    '''.replace('{title}',title)

    for i in range(len(issues)):
        temp_data = temp_data + issues[i].text + ','

    temp_data = temp_data + '\n\nDescriptions\n'
    for i in range(len(issues)):

        temp_data = temp_data + title + ' due to '
        temp_data = temp_data + issues[i].text

        paragraphs = issues_desc[i].find_all('p')
        for p in paragraphs:
          temp_data =  temp_data +'\n' +  p.get_text()

        paragraphs = issues_desc[i].find_all('ol')
        for p in paragraphs:
          temp_data = temp_data + p.get_text()
        temp_data =  temp_data +'\n\n\n'

    # print(temp_data)
    refrigerator_issues_data = refrigerator_issues_data + temp_data
    # break
```

```python
from typing import Literal

from langchain_core.prompts import ChatPromptTemplate, PromptTemplate
from langchain_core.pydantic_v1 import BaseModel, Field
from langchain_openai import ChatOpenAI

from langchain_core.runnables import RunnableLambda
```

```python


# Data model
class DataRouteQuery(BaseModel):
    """Route a user query to the most relevant datasource."""

    datasource: Literal["product_database_lookup", "troubleshooting_guidelines", "compatability_questions","general"] = Field(
        ...,
        description='''Given a user question choose which datasource would be most relevant for answering their questions
                      here product_database_lookup has information about part id, product id, how to install products, all installation guidelines etc. and
                      troubleshooting  guidelines helps in solving various issues about dishwashers and refrigerators and
                      compatability_questions answers weather a part is compatable with the model and
                      for any general questions or greetings should go to general
                      Only answer questions for Dishwasher and Refrigerators all other things do not answer ''',
    )

# LLM with function call
llm = ChatOpenAI(model="gpt-4o", temperature=0.7)
structured_llm = llm.with_structured_output(DataRouteQuery)

# Prompt
system = """You are an expert at routing a user question to the appropriate data source and a kind chatbot for e-commerce site Part Select

Based on the help user is asking for route it to the relevant data source. """

# prompt = ChatPromptTemplate.from_messages(
#     [
#         ("system", system),
#         ("human", "{question}"),
#     ]
# )

prompt = PromptTemplate.from_template(

''' System : {system}

Human_user, {question}


Also smartly make use of past conversations: if they exist {memory} to make decision
                                      '''


)


# Define router
data_router = prompt | structured_llm
```

```python
def choose_data_route(result):
    if "product_database" in result.datasource.lower():
        ### Logic here
        ### Add function for json
        return "chain for products"
    elif "troubleshooting_guidelines" in result.datasource.lower():
        ### Logic here
        ### Add troubel shooting model
        return "chain for troubleshooting"
    elif "compatability_questions" in result.datasource.lower():
        return
    else:   #general
        return 'Please ask question again. Thank You!!'
```

```python
# Define the metadata extraction function.
def metadata_func_dishwasher(record: dict, metadata: dict) -> dict:

    metadata["product_id"] = record.get("PartSelect_Number")
    metadata["manufacture_part_id"] = record.get("Manufacturer_Part_Number")
    metadata['device'] = 'Dishwasher'
    metadata['name'] = record.get("Product_Name")
    metadata['price'] = record.get("Price")

    return metadata
def metadata_func_refrigerator(record: dict, metadata: dict) -> dict:

    metadata["product_id"] = record.get("PartSelect_Number")
    metadata["manufacture_part_id"] = record.get("Manufacturer_Part_Number")
    metadata['device'] = 'Refrigerator'
    metadata['name'] = record.get("Product_Name")
    metadata['price'] = record.get("Price")

    return metadata
```

```python
#Generation
from langchain_openai import ChatOpenAI
from langchain.prompts import ChatPromptTemplate

# Prompt
template = """You are an expert an extracting product_id (they usually start with 'PS' but not always) and model_number to check compatability between them
BOTH CANNOT BE SAME
Given a {query} ONLY extract 'product_id' and 'model_number' in  JSON format.

ONLY IF BOTH EXIST RETURN JSON AND SAY NOTHING

If ONLY ONE value is present kindly ask user for the missing value, DO NOT ASSUME

You can make use of {memory} for past conversation if exists. DO NOT ASK USER FOR CONTEXT

"""
prompt = ChatPromptTemplate.from_template(template)
prompt
```

```python
memory = 'Past Conversation: Is this part compatible with my WDT780SAEM1 model? --> Sure, I can help with that. Please provide the product_id for the part you are inquiring about.'
```
